### Tensorflow puro

```python
model = Sequential()
model.add(InputLayer(input_shape=(13))) #"las dimensiones de los ejemplos"
model.add(Dense(units=20, activation = "relu")) ## units es el numero de neuronas de la capa
model.add(Dense(15, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1, activation= "linear" )) # capa de salida del modelo

model.compile(optimizer=SGD(learning_rate=0.001), loss=MeanSquaredError())
```

```python
def get_weight(shape, name):
    return tf.Variable(initializer(shape), name=name, trainable=True, dtype=tf.float32)
```

```python
shapes = [
    (13, 20), ## primer capa
    (20,15),
    (15,10),
    (10,1) 
]
  
shapes_biases= [
    (20,1), ## primer capa
    (15,1),
    (10,1),
    (1,1) 
]
```

```python
initializer= tf.initializers.glorot_uniform() ## Recordamos que los pesos deben ser aleatorios para evitar simetria
```

```python
weights = []
for i,shape in enumerate(shapes):
    weights.append(get_weight(shape, f"weight{i}"))
    
biases = []
for i,shape in enumerate(shapes_biases):
    biases.append(get_weight(shape, f"bias{i}"))
    
```

```python
def model(x, weights, biases):
    x = tf.cast(x, dtype=tf.float32)
    z = tf.add((tf.matmul(tf.transpose(weights[0]),x)), biases[0])
    a = tf.nn.relu(z)
    
    z2 =  tf.add((tf.matmul(tf.transpose(weights[1]),a)), biases[1])
    a2 = tf.nn.relu(z2)
    
    z3 =  tf.add((tf.matmul(tf.transpose(weights[2]),a2)), biases[2])
    a3 = tf.nn.relu(z3)
    
    z4 =  tf.add((tf.matmul(tf.transpose(weights[3]),a3)), biases[3])
    return z4
```

```python
xtr.shape
```

```python
iterations = 1000
optimizer = tf.optimizers.SGD(learning_rate=0.001)
losses = list()
val_loss= list()

for i in range(iterations):
    with tf.GradientTape() as tape:
        y_pred = model(xtr.T, weights, biases)
        current_loss = tf.losses.mean_squared_error(ytr.reshape(1,-1),y_pred) 
    

    print("iteration {} losss {}".format(i,current_loss.numpy()[0]))
    losses.append(current_loss.numpy()[0])
    grads = tape.gradient(current_loss, weights+biases) ## concatenamos las listas de los pesos
    val_loss_ = tf.losses.mean_squared_error(yvr.reshape(1,-1),model(xvr.T, weights, biases)) 
    val_loss.append(val_loss_)
    optimizer.apply_gradients(zip(grads, weights+biases)) 
    
```
