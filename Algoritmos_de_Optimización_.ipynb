```python
import numpy as np
import plotly.graph_objects as go
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X = np.arange(10, step= .2)
Y = np.sin(X)

Y = Y + np.random.normal(0,.3, Y.shape[0])
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=X,y=Y, name="serie de tiempo"))
fig.show()
```

### Suavizamiento Exponencial

$$V_0 = 0      $$
$$V_1 = \beta V_0+ (1-\beta)Y_1   $$
$$V_t = \beta V_{t-1}+ (1-\beta)Y_{t}  $$

$$V_t^{corrected} = \frac {V_t} {1-\beta^t}$$


```python
beta = .6
suavizado = dict(v0=0)

for i,value in enumerate(Y):
    suavizado[f"v{i+1}"] = beta*suavizado[f"v{i}"] + (1-beta)*value
    suavizado[f"v{i+1}"] = suavizado[f"v{i+1}"] / (1-beta**(i+1))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=X,y=Y, name="serie de tiempo"))
fig.add_trace(go.Scatter(x=X,y=list(suavizado.values())[1:], name="suavizado"))
fig.show()
```

### Dataset

```python
X,Y = make_moons(600, noise=.2)
```

```python
X.shape, Y.shape
```

```python
plt.scatter(X[:,0],X[:,1], c = Y, s=40, cmap = plt.cm.Spectral )
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
X.shape
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(2)))
model.add(Dense(5, activation="linear"))
model.add(LeakyReLU(.3))
model.add(Dense(3, activation="relu"))
model.add(Dense(1,activation="sigmoid"))

model.compile(loss=BinaryCrossentropy(), optimizer = SGD(.001), metrics=["accuracy"])
```

```python
history = model.fit(X,Y, epochs=3000, batch_size=128)
```

```python
from tensorflow.keras.optimizers import Adam
```

```python

model.compile(loss=BinaryCrossentropy(), optimizer = Adam(learning_rate=0.001,
    beta_1=0.9, # La beta de momentum
    beta_2=0.999), # la beta de RMSProp
    metrics= ["accuracy"]        
             )
```

```python
history_adam = model.fit(X,Y, epochs=3000, batch_size=128)
```

### Funcion de perdida

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["loss"]))),y=history.history["loss"], name="SGD"))
fig.add_trace(go.Scatter(x=list(range(len(history_adam.history["loss"]))),y=history_adam.history["loss"], name="Adam"))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["accuracy"]))),y=history.history["accuracy"], name="SGD"))
fig.add_trace(go.Scatter(x=list(range(len(history_adam.history["accuracy"]))),y=history_adam.history["accuracy"], name="Adam"))
fig.show()
```
