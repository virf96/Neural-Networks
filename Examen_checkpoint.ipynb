**1. ¿Qué es una red neuronal?**

Las redes neuronales son modelos computacionales, inspirados de alguna forma en el comportamiento del cerebro humano. Generalmente la red neuronal es un conjunto de nodos llamado neuronas artificiales las cuales están conectados entre si para propagar los datos que ingresan.

Estas redes aprenden a partir de ejemplos para después poder generalizar   su experiencia a nuevos ejemplos y abstraen las características principales de una serie de datos.

Una red neuronal puede modelar comportamientos complejos no lineales y no tienen ninguna restricción en los tipos de datos que se puedan usar, además de que en muchas ocasiones son más eficientes a la hora de predecir, pero no son tan fáciles de explicar

**2. ¿Por qué se agregan neuronas y capas a una red neuronal?**

 El elemento básico y esencial dentro de una red neuronal es la neurona, ya que en ellas se procesa la información, pero ,mientras una neurona es muy pequeña en sí misma, cuando se combinan cientos, miles o millones de ellas pueden resolver problemas muy complejos, es por eso que se agregan varias o cientos de ellas. 

En general las neuronas se suelen agrupar en unidades estructurales que se denominan capas, esto para que el procesamiento de la información tenga más precisión y se consideren la mayoría de las características posibles de los datos y se obtenga un mejor resultado, además de que le da la oportunidad de aprender problemas más difíciles si hay mas de una, hablando de las ocultas.

**3. ¿Qué es es la función de pérdida?**

La función de pérdida tiene la labor de evaluar la desviación entre los valores estimados por la red neuronal y los valores reales de las observaciones utilizadas durante el aprendizaje. En otras palabras es la función que nos va a decir cuánto nos estamos equivocando en el cálculo. 

Un resultado alto significa que la red neuronal tiene un desempeño pobre y un resultado bajo indica que la red neuronal está haciendo un buen trabajo. Esta es la función que optimizamos o minimizamos cuando se realiza el back propagation.

Se denota como:

 $L(a,y)$ (en el caso de un registro) y $L(a^{i}, y^{i})$ (en el caso de varios registros) y generalmente como:

$L(\hat{y}, y)$ = $-(y\ln(\hat{y})+(-y) \ln(- \hat{y}))$

**4. ¿Qué es la función de costo?´**

La función de coste o costo trata de determinar el error entre el valor estimado y el valor real, con el fin de optimizar los parámetros de la red neuronal.

Esta función dependerá de W y de b porque son los parámetros que se buscan optimizar en la red neuronal. 

**5. Describa que es Forward and back propagation.**

El entrenamiento de una red neuronal se divide en dos partes. De la propagación hacia delante ( forward propagation) y propagación hacia atrás ( back propagation).

La propagación hacia delante ( forward propagation), es el proceso en el cual las redes neuronales crean las predicciones y básicamente consiste en el cálculo de la $\hat{y}$. El proceso esta dado por: 

  $a)$ Entran las características $(x1,x2,…,xn)$

  $b)$ Se multiplican por los pesos $( w1,w,2,wn)$

  $c)$ Se les suma el sesgo $(b)$

Denotando matemáticamente a $Z$ = $W^T X + b$
  
  $d)$ Se calcula la función de activación $(a)$, para obtener $\hat{y}$

Denotando matemáticamente $a$= $\sigma (z)$

  $e)$ Se calcula la función de perdida 

Denotando matemáticamente $L(a,y)$ (en el caso de un registro) y $L(a^{i}, y^{i})$ (en el caso de varios registros)

$f)$ Más específicamente para encontrar la función de todos los registros se calcularía la función de coste , que va a ser el promedio de la suma de todas las funciones de perdida. Esta función dependerá de $W$ y de $b$ porque son los parámetros que se buscan optimizar en la red neuronal. 

Se plasma matemáticamente como :

$J(W,b)$ = $\frac{1}{m}$ $\sum_{1=1}^{m} L(a^{i}, y^{i})$ 


Este se repite varias veces según se requiera y en cada ciclo se ejecuta el algoritmo de backpropagation para actualizar los valores de $W$ y $b$.

**La propagación hacia atrás ( back propagation).**

Consiste en el proceso de optimización de los parámetros ($W$ y $b$) para realizar la predicción acertada.  Este algoritmo se encarga de calcular las derivadas (o gradientes) de los parámetros $w$ y $b$ conforme pasan las iteraciones  (o epochs).para saber cómo estos parámetros afectan al resultado de la función de perdida
La idea general es encontrar el minimo global de la función de perdida, lo que hace el algoritmo es descender por la función hasta llegar a este punto donde la función se encuentra optimizada y las derivadas son las que se encargan de encontrar ese camino al punto.

**6. ¿Como se optimizan los parámetros de una red’**

Los parámetros que se buscan optimizar dentro de una red neuronal son ($w$ y $b$), es decir los pesos y los sesgos ya que de ellos depende encontrar el mínimo dentro de la función de costo.

Esos parámetros se optimizan moviendo la tasa de aprendizaje es decir $\alpha$ y eso se da cuando se actualizan los pesos, esta tasa dirá que tan grande va a ser el salto que se va a dar en la función, entre cada iteración , esto en función de llegar al mínimo de la función de coste, para ello también se considera derivar parcialmente respecto a $w$ y a $b$ en función de $J$.

Matemáticamente se actualizaría de la forma:

$w$ : $w-\alpha$ $\frac{\partial J}{\partial w}$

$b$ : $b-\alpha$ $\frac {\partial J} {\partial b}$

**7. Describa qué son y para qué sirven las capas ocultas, la capa de entrada y la capa de salida.**

Dentro de una red , al menos se contará con 3 capas, una de entrada, una oculta y una de salida.

- En la capa de entrada se reciben las características o variables informativas

- Las capas ocultas contienen las neuronas, estas beuronas pueden estar interconectadas de distintas maneras, además de que  gracias a ellas el algoritmo puede aprender relaciones más dificiles de modelar como pueden ser las no lineales gracias a las diferentes funciones de activación que se pueden utilizar en las diferentes capas ocultas que puede llegar a tener una red neuronal.

- La capa de salida es aquella que va a transferir la información de la red , es decir representará la información aprendida a tráves de las predicciones , es  $\hat{y}$.


**8. ¿Por qué se usan funciones de activación?**

Las funciones de activación agregan el componente no lineal dentro de la red, es decir ayudan a que no se caiga en un modelo  lineal, y que cada neurona este calculando algo distinto, además las funciones de activación entre capas ocultas no son lineales solo es lineal cuando se utiliza una sola capa oculta con una neurona.

Al usar diferentes funciones de activación en las diferentes capas ocultas el modelo puede aprender otro tipo de comportamientos más complejos dado que que esto le da la capacidad de modelar relaciones no lineales que son más complejas.


Una función de activación calcula el estado de actividad de una neurona; 
transformando la entrada global menos el umbral, en un valor (estado) de activación, cuyo rango normalmente va de (0 a 1).Tomando el valor de 0 cuando esta esta inactiva es decir no se superó el umbral o 1 en el caso de que si lo haya superado.  Esta función de activación se asigna de acuerdo al tipo de problema que se este trabajando. En el caso de las funciones de clasificación binaria se usaría la función “Sigmoide”, para el caso de un problema de regresión una lineal en el caso de un multiclase una softmax, etc. 
Generalmente se agregan ya que las funciones de activación agregan el componente lo lineal dentro de la red

**9. ¿Por qué no es buena idea inicializar los pesos con ceros en redes profundas?**

Porque se corre con el riesgo de que el modelo no sea funcional, ya que existirá un problema de simetría en las matrices de pesos $W$, ya que las entradas en muchos casos coincidirían es por eso que se recomienda inicializar en números aleatorios, pero no en cero.

**10. Describa en qué se diferencian Batch Gradient Descent, Mini Batch Gradient Descent y Stochastic Gradient Descent.**

**Batch Gradient Descent:**  a comparación con el “Stochastic Gradient Descent”, esta oscila menos y la función de perdida se ve más recta , y se acerca más rápido al mínimo. Aquí se usa el data set completo. 

**Mini Batch Gradient:** Se usa cuando tenemos un Batch más grande que uno solo, es decir : 2, 4, 6, 8, 10, 34. Consiste en pasar por batches en vez de todo el data set completo. Converge más rápido y oscila menos que en el “Stochastic Gradient Descent”

**Stochastic Gradient Descent:** Cuando se actualizan los pesos va a tomar distintos mínimos entonces la función de perdida va a ir oscilando porque puede que la función de perdida en cada iteración sea distinta, y normalmente no converge al mínimo y no tendremos predicciones tan concretas. Se usa para 1.

**11. Describa cómo se agrega una regularización L2 a un red neuronal(la forma teórica, no la forma de implementarlo en tensorflow. )**

El error de la red neuronal se representa con la función de perdida. Y la de coste como el promedio de las funciones de perdida.

Para regular en L2 , es necesario añadir ciertas cosas a ala función de perdida que se denotaba por:

$J(W,b)$= $\frac {1}{m}$ $\sum_{i=0}^{m}$ $L(\hat{y}_{i}, y^{i})$ A la cual se le agregará el siguiente factor:

Lamda, que va a representar el parámetro de regularización , el cual mientras más grande sea lamda mas van a penalizar los pesos y mientras chicos menos penalización van a tener.
Si lo que se busca es minimizar la función , también se considera penalizar los pesos de tal forma que al aplicarle la norma estos se hagan mas chicos. 
Eso para el caso de una capa, en dado caso de tener mas capas lo que se procede a hacer es sumar la norma de la matriz de los pesos en l2, y se le llamaría $L2$ con múltiples capas. Al suceder esto, forzamos al modelo a tener un comportamiento más lineal y esto en situaciones de sobre ajustes nos puede llevar a reflejar mejores resultados.

```python
trainX,trainY = np.array(train_set["train_set_x"]),np.array(train_set["train_set_y"])
testX,testY = np.array(test_set["test_set_x"]),np.array(test_set["test_set_y"])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense,Dropout,MaxPool2D,Conv2D # Para volumenes 
from tensorflow.keras import Input
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1_l2
```

```python
trainX_es = trainX.reshape(trainX.shape[0],64,64,3)
testX_es = testX.reshape(testX.shape[0],64,64,3)
```

```python
model = Sequential()
model.add(Input(shape=(64, 64, 3)))
model.add(Conv2D(8,kernel_size=(6,6),strides=1, padding="same",activation="tanh" ,kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Conv2D(11, kernel_size=(9,9), strides=1, padding="valid", activation="relu",kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten()) # Convierte el volumen en un vector
model.add(Dense(7, activation="relu",kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(Dropout(0.3))
model.add(Dense(10, activation="relu",kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(Dense(units=1, activation="sigmoid"))
```

```python
model.compile(loss=BinaryCrossentropy(),optimizer=Adam(), metrics=["accuracy"])
```

```python
model.summary()
```

```python
hist = model.fit(trainX_es,trainY, batch_size=30,epochs=50, validation_data=(testX_es,testY), verbose = 1)
```

```python
fig = go.Figure(layout = dict(title="Loss"))
fig.add_trace(go.Scatter(x=list(range(len(hist.history["loss"]))),y=hist.history["loss"], name="loss"))
fig.add_trace(go.Scatter(x=list(range(len(hist.history["val_loss"]))),y=hist.history["val_loss"], name="val loss"))
fig.show()
```

```python
from sklearn import metrics

predicted_svm2 = model.predict_classes(testX_es)
print("Accuracy Score Test -> ", metrics.accuracy_score(testY.reshape(-1),predicted_svm2))
#print(text_clf_svm2.score(X_train, y_train))

yr_train_pred =  model.predict_classes(trainX_es)
print ("Accuracy Score Train -> ", metrics.accuracy_score(trainY.reshape(-1), yr_train_pred))

print(metrics.classification_report(testY.reshape(-1), predicted_svm2))
```

```python
def linear(z):
    return z 

def initialize_with_zeros(dim):
    W = np.zeros((1,dim))
    b = 0.
    return W,b

def propagate(W,b,X,Y):
    m = X.shape[1]
    A=linear(np.matmul(W,X)+b)
    cost= np.sum(np.square(Y-A))/m
    cost = np.squeeze(cost)
    dw = np.matmul(X,2*(A-Y).T)/m
    db = np.sum(2*(A-Y))/m
    
    grads = {"dw":dw,"db":db}
    return grads,cost

def optimize(W,b,X,Y, num_iterations, learning_rate= 0.001):
    costs=[]
    for i in range(num_iterations):
        grads, cost = propagate(W,b,X,Y)
        
        dw = grads["dw"]
        db = grads["db"]
        
        W += -learning_rate*dw.T
        b += -learning_rate*db
        
        if i %100 == 0:
            costs.append(cost)
            print("Costo despues de la iteracion %i %.6f"%(i,cost))
            
    params = {"W":W,
               "b": b}
    
    grads = {"dw":dw,
             "db":db}
    return params, grads,costs

def predict(W,b, X):
    A = linear(np.matmul(W,X)+b)
    return A
    

def model(X,Y, num_iterations=10000, learning_rate=.0002):
    W,b = initialize_with_zeros(X.shape[0])
    
    parameters, grads, costs = optimize(W,b,X,Y, num_iterations, learning_rate)
    
    W = parameters["W"]
    b = parameters["b"]
    
    Y_prediction = predict(W,b,X)
    
    d = {"costs":costs,
        "Y_predictions": Y_prediction,
        "W":W,
        "b":b,
        }
    return d
    
```
