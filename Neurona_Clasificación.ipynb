```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

 m = Numero observaciones en nuestro dataset, en este caso es 569 \
 nx = numero de variables de nuestro dataset

```python
def initialize_with_zeros(nx):
    W = np.zeros((1,nx))
    b = 0.0
    return W,b 
```

```python
def propagate(W, b, X, Y):
    z = np.matmul(W,X)+ b
    A = sigmoid(z)
    m = A.shape[1]
    cost = -np.mean((np.multiply(Y,np.log(A))) + np.multiply(1-Y, np.log(1-A)))
    cost= np.squeeze(cost)
    dJ_dW = (1/m) * np.matmul(X,(A-Y).T)
    dJ_db = (1/m) * np.sum(A-Y)
    
    
    return cost, dJ_dW, dJ_db
```

```python
def predict(W,b, X):
    z = np.matmul(W,X)+ b
    A = sigmoid(z)
    return np.where(A>=.5, 1, 0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def model(X,Y, iterations, learning_rate):
    W,b = initialize_with_zeros(X.shape[0])
    
    costs = list()
    accuracies = [accuracy_score(Y,predict(W,b,X))]
    for i in range(iterations):
        cost, dJ_dW, dJ_db = propagate(W,b,X,Y)
        
        costs.append(cost)
        W = W - learning_rate * dJ_dW.T
        b = b - learning_rate * dJ_db
        ## Accuracy 
        predict_classes = predict(W,b,X)        
        accuracies.append(accuracy_score(Y.reshape(-1,1), predict_classes.reshape(-1,1)))
        print("Cost after iteration {}".format(i), costs[i], "Accuracy",accuracies[i] )
    
    return W,b, costs, accuracies
```

```python
W,b, costs, accuracies = model(X,Y, iterations= 10000, learning_rate=1)
```
