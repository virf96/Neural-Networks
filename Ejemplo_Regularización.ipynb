```python
from scipy.io import loadmat
import plotly.graph_objects as go
import numpy as np
```

```python
data = loadmat("data.mat")
```

```python
data.keys()
```

```python
xt = data["X"]
yt = data["y"]
xv= data["Xval"]
yv = data["yval"]
```

```python
xt.shape
```

```python
np.unique(yt)
```

### Red Neuronal

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import Dense, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
xt.shape
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(2)))
model.add(Dense(5, activation="relu"))
model.add(Dense(10, activation="relu"))
model.add(Dense(1,activation="sigmoid"))

model.compile(loss=BinaryCrossentropy(), optimizer = Adam(learning_rate=0.001), metrics=["accuracy"])
```

```python
history = model.fit(xt,yt, epochs=1000, batch_size=128, validation_data=(xv,yv))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["loss"]))),y=history.history["loss"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_loss"]))),y=history.history["val_loss"], name="train"))

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["accuracy"]))),y=history.history["accuracy"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_accuracy"]))),y=history.history["val_accuracy"], name="val"))

fig.show()
```

### Con regularizacion 

```python
from tensorflow.keras.regularizers import l1_l2
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(2)))
model.add(Dense(5, activation="relu", kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(Dense(10, activation="relu", kernel_regularizer = l1_l2(0.001, 0.01)))
model.add(Dense(1,activation="sigmoid", kernel_regularizer = l1_l2(0.001, 0.01)))

model.compile(loss=BinaryCrossentropy(), optimizer = Adam(learning_rate=0.001), metrics=["accuracy"])
```

```python
history = model.fit(xt,yt, epochs=1000, batch_size=128, validation_data=(xv,yv))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["loss"]))),y=history.history["loss"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_loss"]))),y=history.history["val_loss"], name="val"))

fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["accuracy"]))),y=history.history["accuracy"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_accuracy"]))),y=history.history["val_accuracy"], name="val"))

fig.show()
```

### Dropout

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(2)))
model.add(Dense(5, activation="relu",))
model.add(Dense(10, activation="relu", ))
model.add(Dropout(.1))
model.add(Dense(1,activation="sigmoid",))

model.compile(loss=BinaryCrossentropy(), optimizer = Adam(learning_rate=0.001), metrics= ["accuracy"])
```

```python
history = model.fit(xt,yt, epochs=1000, batch_size=128, validation_data=(xv,yv))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["loss"]))),y=history.history["loss"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_loss"]))),y=history.history["val_loss"], name="val"))
fig.update_layout(dict(title="Funcion de perdida"))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=list(range(len(history.history["accuracy"]))),y=history.history["accuracy"], name="train"))
fig.add_trace(go.Scatter(x=list(range(len(history.history["val_accuracy"]))),y=history.history["val_accuracy"], name="val"))
fig.update_layout(dict(title="Accuracy"))
fig.show()
```
